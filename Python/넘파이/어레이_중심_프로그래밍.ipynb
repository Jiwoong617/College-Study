```python
# 넘파이
import numpy as np

# 램덤 시드
np.random.seed(12345)

# 어레이 사용되는 부동소수점들의 정확도 지정
np.set_printoptions(precision=4, suppress=True)

# 파이플롯
import matplotlib.pyplot as plt

# 도표 크기 지정
plt.rc('figure', figsize=(10, 6))
```

### 유니버셜 함수

```python
arr = np.arange(10).reshape((2,5))
arr
```

np.sqrt() : 제곱근 반환

```python
np.sqrt(arr)
```

np.exp() : 지수승 반환

```python
np.exp(arr)
```

np.maximum() : 항목별 최대값 반환

```python
a = np.random.random(10).reshape((2,5))
b = np.random.random(10).reshape((2,5))
print(a)
print(b)
```

```python
np.maximum(a,b)
```

np.divmod() : 나눗셈의 몫의 어레이, 나머지 어레이의 튜플 반환
+ 부동소숫점에 대해서도 실행
+ 두 어레이의 나눗셈도 가능

```python
div = np.arange(1,50,5.1).reshape(2,5)
div2 = np.arange(1,11).reshape(2,5)
```

```python
np.divmod(div,3)
```

```python
np.divmod(div, div2)
```

np.meshgrid() : 지정된 1차원 어레이 2개를 이용하여 격자무늬 좌표생성

```python
dot = np.arange(-1, 1.1, 0.1)
```

```python
a, b = np.meshgrid(dot,dot)
a
```

```python
plt.rc('figure', figsize=(6,6))#도표 크기지정

plt.scatter(a,b)
plt.show()
```

```python
z = np.sqrt(a**2 + b**2)
z.shape
```

```python
plt.rc('figure', figsize=(10, 6))

plt.imshow(z, cmap=plt.cm.gray, extent=[-1,6, 6, -1])
plt.colorbar() #컬러바 표기
plt.title("hole")
#plt.scatter(z,z)
plt.show()
```

### 통계
**axis : 축**
* axis = 0 : 행을 기준으로 함수 적용
* axis = 1 : 열을 기준으로 함수 적용

```python
arr = np.arange(1,11).reshape(2,5)
arr
```

mean() : 어레이에 포함된 모든값들의 평균 계산

```python
arr.mean(axis = 0)
```

```python
np.mean(arr, axis = 0)
```

sum() : 어레이에 포함된 모든값들의 합

```python
arr.sum()
```

```python
np.sum(arr)
```

cumsum() : 어레이의 포함된 모든값들의 누적합

```python
arr.cumsum()
```

```python
np.cumsum(arr)
```

cumprod() : 어레이에 포함된 모든값들의 누적곱

```python
arr.cumprod()
```

```python
np.cumprod(arr)
```

### 부울 어레이


```python
arr = np.random.randn(100)
arr
```

```python
(arr>0).sum()
```

```python
ar = np.array([True, False, False, True])
```

any() : 부울 어레이에 포함된 항목이 하나라도 참이면 참 반환

```python
np.any(arr>0)
```

```python
ar.any()
```

all() : 부울 어레이에 포함된 항목이 모두 참일 때만 참

```python
np.all(arr>0)
```

```python
ar.all()
```

where(조건, 참값, 거짓값) : 조건 검사후 참이면 참값 반환, 거짓이면 거짓값 반환
- 참값, 거짓값을 안써주면 인덱스 반환

```python
np.where(arr>0)
```

```python
arr1 = np.where(arr>0, 1, 0)
arr1
```

### 정렬

```python
arr = np.random.randn(5, 3)
arr
```

sort() : 정해진 축을 기준으로 오름차순 정렬, 축 정하지 않으면 열을 기준
- 행: 0, 열 : 1
- 기존 어레이를 직접 변환

```python
arr.sort()
arr
```

```python
arr.sort(0)
arr
```

argmax() : 지정된 축을 기준으로 항목들의 가장 큰값이 위치한 인덱스 반환

argmin() : 지정된 축을 기준으로 항목들의 가장 작은값이 위치한 인덱스 반환
- 축지정 안할 시 전체를 기준으로 함

```python
arr = np.random.randint(11, 99, (2,5))
arr
```

```python
arr.argmax(1)
```

```python
arr.argmax(0)
```

```python
arr.argmin()
```

```python
arr.argmin(0)
```

### 난수 생성

np.random.randn() : 임의의 부동소수점을 표준 정규 분포를 따르면서 지정된 수만큼 생성

```python
np.random.randn(10)
```

```python
np.random.randn(2,5)
```

np.random.rand() : 0과 1사이의 임의의 부동소수점을 균등분포를 따르며 지정된 수만큼 생성

```python
np.random.rand(10)
```

```python
np.random.rand(2,3)
```

np.random.randint() : 지정된 구간 사이에서 정수를 균등분포를 따르며 지정된 수만큼 생성

```python
np.random.randint(5, 10, 10)
```

```python
np.random.randint(5,10, (2,3))
```
