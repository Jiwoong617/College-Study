### 어레이 모양 변형

**reshape()** : 지정된 튜플의 모양으로 주어진 어레이 모양 변형
* 항목의 수만 같으면  임의 차원의 어레이를 임의 차원의 어레이로 변경 가능
* -1 사용시 값이 자동으로 지정됨.

```python
arr = np.arange(8).reshape(2,4)
arr
```

```python
arr.reshape(2,2,2)
```

```python
arr.reshape(4,-1)
```

### 차원 추가

```python
a = np.random.rand(5)
a
```

```python
a[:, np.newaxis] #reshape(5,1)과 동일
```

```python
a[np.newaxis, :] #reshape(1,5)와 동일
```

**ravel() 과 flatten()** : 1차원 어레이 반환
* ravel()메서드는 view를 사용하여 반환된 어레이 변경시 원래 어레이도 같이 변경
* flatten() 메서드는 어레이를 새로생성

```python
arr = np.arange(15).reshape(5,3)
```

```python
arr1 = arr.ravel()
print(arr1)
arr1[0] = 99
print(arr) # arr값이 변경됨
```

```python
arr2 = arr.flatten()
arr2
```

### 어레이 분열/결합

```python
x = np.arange(20).reshape(4,5)
x
```

**axis 축**
+ axis = 0 -> 행 (열 방향/ 수직)
+ axis = 1 -> 열 (행 방향/ 수평)

```python
print(x.sum(axis=0))
print(x.sum(axis=1))
```

np.split(행렬, [인덱스], axis) : 지정된 기준에 따라 쪼갬

```python
np.split(x, [2,3], axis=0)
```

```python
np.split(x, [1,3], axis =1)
```

**np.vsplit()/np.hsplit()**
* vsplit은 axis = 0
* hsplit은 axis = 1

```python
np.vsplit(x, [2,3]) # np.split(x, [2,3], axis=0)와 같음
```

```python
np.hsplit(x,[1,3]) #np.split(x, [1,3], axis =1)와 같음
```

np.concatenate() : 두개의 어레이를 이어붙인다.
- axis=0 : 위아래(행)로 이어붙이기
- axis=1 : 좌우(열)로 이어붙이기

```python
arr1 = np.array([[1, 2, 3], [4, 5, 6]])
arr1
```

```python
arr2 = np.array([[7, 8, 9], [10, 11, 12]])
arr2
```

```python
np.concatenate([arr1, arr2], axis=0)
```

```python
np.concatenate([arr1, arr2], axis=1)
```

**np.vstack()/np.hstack()**
* vstack은 axis = 0
* hstack은 axis = 1

```python
np.vstack((arr1, arr2))
```

```python
np.hstack((arr1, arr2))
```

np.r_[ ] = np.vstack()

np.c_[ ] = np.hstack()

```python
np.r_[arr1, arr2]
```

```python
np.c_[arr1, arr2]
```

### 브로드캐스팅

모양이 다른 두 어레이의 연산이 가능한 경우, 브로드캐스팅 작동
- 하나의 정수와 하나의 어레이

```python
arr = np.arange(6).reshape((2,3))
arr
```

```python
arr*4
```

- 열이 같은 1차원 어레이와 2차원 어레이

```python
arr2 = np.arange(4).reshape((4,1)).repeat(3,axis=1)
arr2 #reapeat() : 축에대해 지정된 숫자만큼 반복
```

```python
arr3 = np.arange(1, 4)
arr3
```

```python
arr2 + arr3
```

- 행은 같지만 열이 다른 두 2차원 어레이

```python
arr4 = np.arange(1, 5).reshape((4,1))
arr4
```

```python
arr2 + arr4
```
