**dataframe.groupby(' ')**
- 작은 따옴표 안, 열 이름의 같은 값을 하나로 묶어 통계/집계를 낼때 사용

### 19. 데이터 프레임에 여러개의 함수를 적용한 출력 합치기

````python
orders.head()
````

````python
orders.groupby('order_id').item_price.sum().head()
#groupby로 order_id의 같은 값 하나로 묶고
#item_price열의 값 더해서 5개만 출력
````

````python
len(orders.groupby('order_id').item_price.sum())
#위의 것을 묶고 나온 값의 길이
````

````python
len(orders.item_price) #묶지 않고 길이 출력 했을 시
````

**.transform()**
- transform 메서드 사용 시 gropby로 같은 값이 합쳐졌어도 원래의 각 요소별로 적용
- 기존의 것에서 새로운 것을 추출 함으로 길이 변화 없음

````python
total_price = orders.groupby('order_id').item_price.transform('sum')
#groupby로 같은 값을 묶었어도 transform으로 sum함수 사용 시 길이 변경 없음
len(total_price)
````

````python
orders['total_price'] = total_price
#orders 데이터프레임에 위에서 구한 total_price붙여주기
orders.head()
````

### 21. 데이터 시리즈 재구조화

````python
titanic.groupby(['Sex','Pclass']).Survived.mean()
#타이타닉 성별,pclass을 묶고 묶인 것들의 생존자 평균이 시리즈 형태로 나타남
````

**.stack()/unstack()**
- stack() : 멀티 인덱스 형태의 데이터를 위에서 아래로 재구조화 (행)
- unstack() : 멀티 인덱스 형태의 데이터를 왼쪽에서 오른쪽으로 재구조화 (열)

````python
titanic.groupby(['Sex', 'Pclass']).Survived.mean().unstack()
````

### 22. 피벗 테이블 만들기

**dataframe.pivot_table()**

각 파라미터의 = 뒤에 리스트형태도 들어갈 수 있음.
- index = 행에 들어갈 열
- columns = 열에 들어갈 열
- values = 값에 들어갈 열
- aggfunc = 데이터 집계함수
- margins = True : 모든 데이터 분석 결과를 뒤에 붙이기

````python
titanic.pivot_table(index='Sex', columns='Pclass', values='Survived', aggfunc=['mean','count'],
                    margins=True)
````
