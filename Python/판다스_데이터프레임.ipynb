### 데이터 프레임

```python
series1 = pd.Series([4, 5, 6, 3 , 1], name="Mango")
series1
```

```python
series2 = pd.Series([5, 4, 3, 0, 2], name="Apple")
series2
```

```python
series3 = pd.Series([2, 3, 5, 2, 7], name="Banana")
series3
```

시리즈를 사전형으로 하여 데이터 프레임 생성

```python
dict1 = {series1.name:series1, series2.name:series2, series3.name:series3}
frame1 = pd.DataFrame(dict1)
frame1
```

pd.concat() : 축에 따라 이어붙이기

```python
pd.concat([series1, series2, series3], axis=1)
```

사전형으로 데이터 프레임 만들기

```python
dict2 = {'state': ['Ohio', 'Ohio', 'Ohio', 'Nevada', 'Nevada', 'Nevada', 'NY', 'NY', 'NY'],
         'year': [2000, 2001, 2002, 2001, 2002, 2003, 2002, 2003, 2004],
         'pop': [1.5, 1.7, 3.6, 2.4, 2.9, 3.2, 8.3, 8.4, 8.5]}
frame2 = pd.DataFrame(dict2)
frame2
```

중첩 사전으로 하면 키 내부의 키는 행의 인덱스로 사용

```python
dict3 = {'Nevada': {2001: 2.4, 2002: 2.9},
         'Ohio': {2000: 1.5, 2001: 1.7, 2002: 3.6}}
frame3 = pd.DataFrame(dict3)
frame3
```

dataframe.index.name : 행 이름 지정

dataframe.columns.names : 열 이름 지정

```python
frame3.index.name = 'year'
frame3.columns.name = 'state'
frame3
```

dataframe.values : 항목들로 이루어진 2차원어레이

```python
frame3.values
```

```python
frame2.values
```

### columns와 index

columns 속성을 사용하여 열의 순서 및 열 추가

```python
frame2 = pd.DataFrame(dict2, columns=['year', 'state', 'pop', 'debt'])
frame2
```

dataframe[열] = 값 : 새로운열 추가 및 기본값 수정 가능

```python
frame2['new'] = np.linspace(1,10, 9)
frame2
```

```python
frame2.columns #열 이름확인
```

index 속성 사용하여 행 이름 지정
- columns 와 index 두개 동시에 지정 가능
- 중복도 가능

```python
frame2 = pd.DataFrame(dict2, columns=['year', 'state', 'pop', 'debt'],
                      index=['one', 'two', 'two', 'four',
                             'five', 'six', 'six', 'eight', 'nine'])
frame2
```

### 데이터 프레임 인덱싱

**열 인덱싱**
- 대괄호 및 속성 형식 사용 가능

```python
frame2['state']
```

```python
frame2.state
```

```python
frame2['Ohio state'] = frame2.state == 'Ohio'
frame2
```

```python
frame2['Ohio state'] # 속성형식으로 사용하면 공백때문에 오류
```

**열 삭제**
- del 예약어 사용시 열 삭제 가능
- drop() 사용하여 삭제 -> 원래 값에는 영향을 안미침 고로, 변수 설정해야됨

```python
del frame2['Ohio state']
frame2
```

```python
f2 = frame2.drop(columns='debt') 
f2
```

**행 인덱싱**
- dataframe.loc[행이름] : loc으로 행이름 지정하여 인덱싱
- dataframe.iloc[번호] : 번호로 인덱싱 가능

```python
frame2.loc['four']
```

```python
frame2.loc[['four', 'five']] #여러개 인덱싱시 ㅁ
```

```python
frame2.iloc[5:7]
```

**열 업데이트**

```python
frame2['debt'] = np.arange(9.)
frame2
```
